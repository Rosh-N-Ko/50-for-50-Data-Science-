## Let's use the regex library to seperate the date and time from the users and their texts into a pandas DataFrame  
We have  a lot of words and a repeated pattern of the date and time ,so make that into a regex pattern.

27/01/2020, 10:15 pm - Dheeraj Lalwani (TSEC, CS): Actually great


### Error 1
While trying to find the correct regex expression I ran into errors in the final product where the text would come up
like so:
    pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s(am|pm)\s-\s'
    
    COnsole :
    'Tanay Kamath (TSEC, CS): Here\n',
     'am',
    'Tanay Kamath (TSEC, CS): Hence this is the most appropriate\n',
     'am',
 
Only am or pm seems to be taken in depending on the order and not both , so the solution is to use'\w' to allow for the separation and it works  

```python
pattern='\d{1,2}\/\d{2,4}\/\d{2,4},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s'
type(pattern)
```
